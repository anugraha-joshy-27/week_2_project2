```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data[column] < lower_bound) | (data[column] > upper_bound)
    data = data[~outliers]
    return data
```

```python
numerical_columns = df_train.select_dtypes(include='number').columns
for column in numerical_columns:
    df_train = remove_outliers_iqr(df_train, column)
```

# Phase 2

### KMeans

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

selected_features = df_train[['customer_age', 'balance']]


k = 3  
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(selected_features)
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
df_train['cluster'] = cluster_labels


plt.scatter(selected_features['customer_age'], selected_features['balance'], c=cluster_labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Customer Age')
plt.ylabel('Balance')
plt.legend()
plt.show()
```

```python
cluster_summary = df_train.groupby('cluster').describe()
print("Cluster Summary:")
print(cluster_summary)
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
df_train['cluster'] = agglomerative.fit_predict(df_train[['customer_age', 'balance']])


plt.figure(figsize=(10, 6))
sns.scatterplot(x='customer_age', y='balance', hue='cluster', data=df_train, palette='viridis')
plt.title('Agglomerative Clustering Results')
plt.show()
```

```python
def build_model(X,y):
    X = sm.add_constant(X) 
    lm = sm.OLS(y,X).fit()
    print(lm.summary()) 
    return X
    
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train_rfe,X_train_rfe)
```
